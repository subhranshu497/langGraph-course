```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from IPython.display import Image

class BatterState(TypedDict):
    runs :int
    balls:int
    noofFours: int
    noofSixes: int
    sr:float
    bpb:float
    boundary_percent:float
    summary: str

def calculate_strikeRate(state: BatterState):
    strike_rate = (state['runs']/state['balls'])*100
    
    return {'sr':strike_rate}

def calculate_ballsperBoundary(state: BatterState):
    bpb = state['balls']/(state['noofFours']+state['noofSixes'])


    return {'bpb':bpb}

def calculate_boundary_percentage(state:BatterState):

    boundary_percent = (((state['noofFours']*4) + (state['noofSixes']*6))/state['runs'])*100


    return {'boundary_percent':boundary_percent}

def summary(state:BatterState)->BatterState:

    summary = f"""
        Strike Rate  - {state['sr']} \n
        Balls per Boundary - {state['bpb']}\n
        Boundary percentage - {state['boundary_percent']}
    """

    return {'summary':summary}

graph = StateGraph(BatterState)

#add nodes 

graph.add_node('calculate_strikeRate', calculate_strikeRate)
graph.add_node('calculate_ballsperBoundary', calculate_ballsperBoundary)
graph.add_node('calculate_boundary_percentage', calculate_boundary_percentage)
graph.add_node('summary', summary)

#add edges
graph.add_edge(START, 'calculate_strikeRate')
graph.add_edge(START, 'calculate_ballsperBoundary' )
graph.add_edge(START, 'calculate_boundary_percentage')
# now above three edges will connected to summary
graph.add_edge('calculate_strikeRate', 'summary')
graph.add_edge('calculate_boundary_percentage','summary')
graph.add_edge('calculate_ballsperBoundary', 'summary')

graph.add_edge('summary', END)

workflow = graph.compile()
#To see the graph 
Image(workflow.get_graph().draw_mermaid_png())

initial_state = {
    'runs':100,
    'balls': 50,
    'noofFours':6,
    'noofSixes':4
}

workflow.invoke(initial_state)


```
